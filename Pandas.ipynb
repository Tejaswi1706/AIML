```python
wine_reviews['pointtoprice'] = (wine_reviews.points)/(wine_reviews.price)
wine_reviews.head()
wine_reviews.pointtoprice.max()
wine_reviews.loc[wine_reviews.pointtoprice.idxmax(), 'title']
```

# Map and Apply

```python
import pandas as pd
count_tropical = 0
count_fruity = 0
for series in wine_reviews['description']:
    if 'tropical' in series:
        count_tropical += 1
    if 'fruity' in series:
        count_fruity += 1
        

descriptor_counts = pd.Series([count_tropical, count_fruity])
descriptor_counts
```

```python
n_trop = wine_reviews.description.map(lambda desc: "tropical" in desc).sum()
n_fruity = wine_reviews.description.map(lambda desc: "fruity" in desc).sum()
descriptor_counts = pd.Series([n_trop, n_fruity], index=['tropical', 'fruity'])
descriptor_counts
```

```python
def stars(row):
    if row['points'] >= 95:
            return 3
    elif row['points'] >= 85 or row['points'] < 95:
            return 2
    elif row['points'] < 85:
            return 1
    elif row['country'] == 'Canada':
            return 3

wine_reviews['star_ratings'] = wine_reviews.apply(stars, axis = 'columns')
wine_reviews['star_ratings']
```

# Grouping and Sorting

```python
wine_reviews.groupby('points').points.count()
```

```python
wine_reviews.groupby('winery').apply(lambda d: d.title.iloc[0])
```

```python
wine_reviews.groupby('winery').title.first()
```

```python
wine_reviews.groupby(['winery', 'title']).points.max()
```

```python
best_rating_per_price = wine_reviews.groupby('price').apply(lambda d: d.loc[d.points.idxmax()])
best_rating_per_price
```

```python
best_rating_per_price = wine_reviews.groupby('price').apply(lambda d: d.loc[d['points'].idxmax(), 'points'])
best_rating_per_price
```

```python
price_extremes = wine_reviews.groupby('variety').price.agg([min, max]).sort_values(by='min', ascending = False)
price_extremes
```

```python
country_variety_counts = wine_reviews.groupby(['country', 'variety']).title.count()
country_variety_counts
```
