```python
trip_advisor_csv['Review_Lower'] = trip_advisor_csv['Review'].str.lower()
```

```python
trip_advisor_csv.head()
```

# Remove Stopwords

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

from nltk.stem import PorterStemmer

nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer

import pandas as pd
```

```python
stopword = stopwords.words('english')
trip_advisor_csv['Review_without_stopwords'] \
= trip_advisor_csv['Review_Lower'] \
.apply(lambda x: ' '.join([word for word in x.split() if word not in stopword]))
```

```python
trip_advisor_csv.head()
```

# Removing punctuation

```python
trip_advisor_csv['Review_without_stopwords_no_punct'] \
= trip_advisor_csv['Review_without_stopwords'].apply(lambda x: re.sub(r"\*", "star", \
                                                                      x))
```

```python
trip_advisor_csv.head()
```

```python
trip_advisor_csv['Review_without_stopwords_no_punct'] \
= trip_advisor_csv['Review_without_stopwords_no_punct'].apply(lambda x: re.sub(r"[^\w\s]", "", \
                                                                      x))
```

```python
trip_advisor_csv.head()
```

# Tokenization

```python
trip_advisor_csv['tokenized'] = trip_advisor_csv['Review_without_stopwords_no_punct'].apply\
(lambda x: word_tokenize(x))
                                                                        
```

```python
trip_advisor_csv.head()
```

# Stemming

```python
ps = PorterStemmer()

trip_advisor_csv['stemmed'] = trip_advisor_csv['tokenized'].apply\
(lambda x: [ps.stem(token) for token in x])
```

```python
trip_advisor_csv.head()
```

# Lemmatization

```python
lm = WordNetLemmatizer()

trip_advisor_csv['lemmatized'] = trip_advisor_csv['stemmed'].apply\
(lambda x: [lm.lemmatize(token) for token in x])
```

```python
trip_advisor_csv.head()
```

# N-grams

```python
tokens = sum(trip_advisor_csv['lemmatized'], [])

print(pd.Series(nltk.ngrams(tokens, 1)).value_counts())
print(pd.Series(nltk.ngrams(tokens, 2)).value_counts())
print(pd.Series(nltk.ngrams(tokens, 3)).value_counts())
```
