### Boolean Indexing or Masking

```python
a = np.array([1, 2, 3, 4, 5])
mask = a > 3
print(mask)
```

```python
print(a[mask])
```

```python
b = np.array([[1, 2, 3], [4, 5, 6]])
mask = b > 4
print(mask)
```

```python
b[mask]
#always returns a flatted 1-d vector
```

### Linear Algebra Essentials

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

a @ b #dot product
```

```python
a.T
```

```python
c = np.array([[1, 2], [3, 4]])
np.linalg.inv(c)
```
